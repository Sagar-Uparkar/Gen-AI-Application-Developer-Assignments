### 4. Problem Statement  

Write a Python program to draw a 3D plot that visualizes the regression model for house price prediction using suitable Python-based 3D plotting libraries. 

Assume the following features were used: 

1. Area (sq ft) 
2. Number of Bedrooms 
3. House Price 

```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
house_dic = {
    "size_sqft":[800, 900, 1200, 1500, 1800, 2000, 2300],
    "bedrooms":[1, 2, 2, 3, 3, 4, 4],
    "price":[3000000, 3500000, 5000000, 6500000, 7500000, 9000000, 11000000]
}

house_df = pd.DataFrame(house_dic)
```

```python
# sample data

house_df.head()
```

```python
# Data preparation

x = house_df[['size_sqft', 'bedrooms']]
y = pd.DataFrame(house_df['price'])

# Fit the model

model = LinearRegression()
model.fit(x, y)
```

```python
# Make prediction

price_pred = model.predict(x)
```

```python
# 3D Plot

fig = plt.figure(figsize = (10, 8))
ax = fig.add_subplot(111, projection = '3d')

ax.scatter(house_df.size_sqft, house_df.bedrooms, house_df.price, color = 'red', label = 'Actual Data', alpha = 0.6)

surf = ax.plot_surface(house_df.size_sqft, house_df.bedrooms, price_pred, color = 'green', alpha = 0.4, label = 'Regression Plane')

ax.set_xlabel('Area (sq ft)')
ax.set_ylabel('Bedrooms')
ax.set_zlabel('Price ($)')
ax.set_title('3D House Price Regression Model')

plt.show()
```
